```r

# Paramètres de la distribution
mu <- 0    # moyenne
sigma <- 2 # écart-type
x <- 1     # valeur de comparaison

# Calculer la probabilité
prob <- pnorm(x, mean = mu, sd = sigma)

# Tracer la distribution
curve(dnorm(x, mean = mu, sd = sigma),
       from = mu - 4*sigma,
       to = mu + 4*sigma,
       main = paste("Distribution Normale N(", mu, ",", sigma, ")"),
       ylab = "Densité de probabilité",
       xlab = "Valeur",
       col = "blue",
       lwd = 2)

# Ajouter une zone colorée jusqu'à x
x_seq <- seq(mu - 4*sigma, x, length = 100)
y_seq <- dnorm(x_seq, mean = mu, sd = sigma)
polygon(c(x_seq, rev(x_seq)),
        c(y_seq, rep(0, length(y_seq))),
        col = rgb(0, 0, 1, 0.3),
        border = NA)

# Ajouter des lignes et annotations
abline(v = mu, col = "red", lty = 2)  # ligne moyenne
abline(v = x, col = "green", lty = 2) # ligne x

# Texte de probabilité
text(x + 0.5, 0.2,
     paste("P(X ≤ ", x, ") =", round(prob, 4)),
     col = "dark green")

# Afficher la probabilité
print(paste("Probabilité P(X ≤", x, ") =", round(prob, 4)))
cat("pnorm(1,0,2)=",pnorm(1,0,2),"\n")
```

```r
sd<-2
# Générer des nombres aléatoires suivant une loi normale
set.seed(123)  # Pour reproduire le même résultat aléatoire
n <- 100000   # Nombre de points
x <- rnorm(n,0,sd)  # Générer n points selon N(0,1)

# Créer un graphique avec transparence
plot(x, rep(0, n),  # Les points sont sur une ligne horizontale
     pch = 16,       # Points pleins
     col = rgb(0, 0, 1, 0.01),  # Bleu avec transparence très faible
     main = "Distribution de points aléatoires selon rnorm()",
     xlab = "Valeur",
     ylab = "",
     yaxt = "n")  # Supprimer l'axe des y

# Ajouter des lignes verticales pour les repères statistiques
abline(v = 0, col = "red", lty = 2)  # Moyenne
abline(v = c(-sd, sd), col = "green", lty = 2)  # ±1 écart-type
abline(v = c(-2*sd, 2*sd), col = "orange", lty = 2)  # ±2 écarts-types
abline(v = c(-3*sd, 3*sd), col = "purple", lty = 2)
```

1 écart-type ($\pm \sigma$) : ~68.27% des données\
2 écarts-types ($\pm 2\sigma$) : ~95.45% des données\
3 écarts-types ($\pm 3\sigma$) : ~99.73% des données

```r

# Paramètres
mu <- 0    # moyenne
sigma <- 2 # écart-type
x <- 1     # valeur de comparaison

# Calculer la probabilité
prob <- pnorm(x, mean = mu, sd = sigma)

# Créer un plot de la CDF
plot(function(p) qnorm(p, mean = mu, sd = sigma),
    from = 0, to = 1,
    main = "Fonction Quantile (Inverse de CDF)",
    xlab = "Probabilité",
    ylab = "Valeur",
    col = "blue")

# Ajouter des lignes verticale et horizontale
abline(v = 0.6915, col = "red", lty = 2)
abline(h = 1, col = "green", lty = 2)

# Ajouter un point pour la probabilité de pnorm(1,0,2)
points(prob, x, col = "red", pch = 16, cex = 1.5)

# Ajouter des annotations
text(prob, x+0.5,
    paste("P(X ≤ 1) =", round(prob, 4)),
    col = "dark green")
```
